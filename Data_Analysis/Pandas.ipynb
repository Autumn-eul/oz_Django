```python
import pandas as pd
```

```python
# Series
data = ['a', 'b', 'c', 'd', 'e']

se = pd.Series(data)

type(se) # pandas.core.series.Series
se.index # RangeIndex(start=0, stop=5, step=1)
se.values # array(['a', 'b', 'c', 'd', 'e'], dtype=object)

se[0:3]

se.name = "alphabet"
se.index.name = "index"
print(se)
```

```python
# DF (DataFrame = Excel)
data = {
    "country" : ["kor", "usa", "china", "japan"],
    "rank" : [1, 2, 3, 4],
    "grade" : ["A", "B", "C", "D"]
}

df = pd.DataFrame(data) # csv to json, json to csv, xlsx
```

```python
# 데이터 셀렉션 => 데이터를 불러오는 방법
# (1) df.컬럼, df['컬럼']

df.grade
df[['country', 'rank', 'grade']]
```

```python
# (2). df.loc[인덱스값, 컬럼명]
df.loc[3]
df.loc[:, ['grade', 'country']] # index 값을 여러개 가져오려면 []로 한번 더 감싸준다

df[df['rank'] > 2]
df.loc[df['rank'] > 2, ['grade', 'country']]
df[df['rank'] > 2][['grade', 'country']] # 위의 코드와 동일한 결과를 출력
```

```python
# 삭제하는 방법 (drop)
# (1) 단순하게 행 데이터를 삭제하는 방법

# df.drop([3]) : 이렇게 하면 삭제가 안되고 사용자가 허락해야 완전히 삭제가 됨
df.drop([3], inplace = True)
df
```

```python
# (2) 컬럼을 삭제
df.drop('rank', axis = 1, inplace = True)
# df = df.drop('rank', axis = 1)
df
```

```python
# 데이터 정렬 => sort_index or sore_values()
df['new_rank'] = [10, 100, 30]

df.sort_values(by = 'new_rank', ascending = False) # 내림차순
```

```python
# null 데이터(NaN) 처리
# isnull() => 데이터의 null 여부
# fillna() => null 데이터 채우기
# dropna() => null 데이터 지우기

import numpy as np
df.loc[1, 'grade'] = np.nan
df.isnull().sum()
```

```python
df['grade'] = np.nan # grade 데이터를 모두 NaN 으로 바꾸고
df.fillna(value = "pass", inplace = True) # pass 라는 값으로 넣어준다
df
```

### 영화 데이터 출력

```python
import pandas as pd

json_data = {
    "columns": ["Movie", "Release Year", "Audience", "Rating"],
    "index": [0, 1, 2, 3, 4, 5, 6, 7],
    "data": [
        ["Avengers", 2012, 1500, 8.8],
        ["Interstellar", 2014, 1100, 9.1],
        ["Frozen", 2013, 1020, 8.5],
        ["About Time", 2013, 950, 8.7],
        ["The Dark Knight", 2008, 1300, 9.0],
        ["Inception", 2010, 1200, 8.8],
        ["La La Land", 2016, 800, 8.6],
        ["Toy Story", 2010, 980, 8.5]
        ]}

# JSON 데이터를 DataFrame으로 변환
df = pd.DataFrame(json_data['data'], columns=json_data['columns'])
df
```

```python
# 1. 전체 데이터 중에서 'Movie' 정보만 출력하시오

df['Movie']

# == df.Movie : 단일 정보 출력
```

```python
# 2. 전체 데이터 중에서 'Movie', 'Rating' 정보를 출력하시오

df[['Movie', 'Rating']]
```

```python
# 3. 2013년 이후 개봉한 영화 데이터 중에서 'Movie', 'Rating' 정보를 출력하시오

df[df['Release Year'] > 2013][['Movie', 'Rating']]

# == df.loc[df["Release Year"] > 2013]
```

```python
# 4. 주어진 계산식을 참고하여 'Recommend' Column 을 추가하시오
# Recommend = (Audience * Rating) // 100

df['Recommend'] = (df['Audience'] * df['Rating']) // 100

df.sort_values(by = 'Recommend', ascending = False).head(5) # top 5
```

```python
# 5. 전체 데이터를 'Release Year' 기준 내림차순으로 출력하시오.

df.sort_values(by = 'Release Year', ascending = False)
```
